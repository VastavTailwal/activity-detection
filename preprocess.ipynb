```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/downstairs.csv')
df2 = pd.read_csv('data/jogging.csv')
df3 = pd.read_csv('data/sitting.csv')
df4 = pd.read_csv('data/standing.csv')
df5 = pd.read_csv('data/upstairs.csv')
df6 = pd.read_csv('data/walking.csv')
df7 = pd.read_csv('data/sleeping.csv')
```

```python
df1['motion'] = 1
df2['motion'] = 1
df3['motion'] = 0
df4['motion'] = 0
df5['motion'] = 1
df6['motion'] = 1
df7['motion'] = 0
```

```python
df1['activity'] = 0
df2['activity'] = 1
df3['activity'] = 2
df4['activity'] = 3
df5['activity'] = 4
df6['activity'] = 5
df7['activity'] = 6
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)
```

```python
df.columns = ['roll', 'pitch', 'yaw', 'gravity_x', 'gravity_y', 'gravity_z', 'rot_x', 'rot_y', 'rot_z', 'acc_x', 'acc_y', 'acc_z', 'motion', 'activity']
```
